### Récupération des ids des articles provenant du Maroc

```python
url = 'https://openapi.etsy.com/v3/application/listings/active?'
# Clé d'API Etsy
api_key = 'p6intkohd066i2oe5pmnafbl' #Idrissa's first api
shared_secret = 'bhceei7yhw'
headers = {    "x-api-key": api_key,
               "x-shared_secret": shared_secret,
               "Content-Type": "application/json"}
params = {    'shop_location': 'Morocco',
              'limit' : 100 }
# Effectuer la requête API
response = requests.get(url, headers=headers, params=params)
# Vérifier si le fichier existe déjà
if os.path.exists('listing_ids.csv'):
    with open('listing_ids.csv', 'r') as f:
        reader = csv.reader(f)
        all_listing_ids = set(row[0] for row in reader)
else:
    all_listing_ids = set()
num_requests = 0
# récupérer les données de toutes les pages
while True:
    response = requests.get(url, headers=headers, params=params)
    num_requests += 1
    data = response.json()
    # extraire les IDs de la page actuelle
    for result in data['results']:
        listing_id = result['listing_id']
        if listing_id not in all_listing_ids:
            all_listing_ids.add(listing_id)
    # sortir de la boucle si toutes les 'id' des produits ont été collectés ou si on atteint le nombre maximal de requêtes
    if len(all_listing_ids) >= data['count'] or num_requests > 2:
        break
    # incrémenter le nombre de résultats à récupérer pour la prochaine page
    params['limit'] = min(100, data['count'] - len(all_listing_ids))
    time.sleep(0.2)
with open('listing_ids.csv', 'a', newline='') as f:  # stocker les IDs récupérées dans le fichier CSV
    writer = csv.writer(f)
    for id in all_listing_ids:
      writer.writerow([id])
```
